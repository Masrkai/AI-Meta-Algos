## Traversal : Visiting without a function

```python

#? Basic Imports
# import networkx as nx
# import matplotlib.pyplot as plt
# from IPython.display import display, Markdown


graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['G'],
    'E': [],
    'F': [],
    'G': []
}
```

```python

#* Breadth First Traversal
def BFTraversal(graph, start):
    if start not in graph:
        return "Start node not in graph"

    visited = []
    queue = [start]  # Use a normal list as a queue

    while queue:
        node = queue.pop(0)  # Pop from the front of the queue (inefficient for large queues)

        if node not in visited:
            visited.append(node)
            adjacent_nodes = graph.get(node, [])
            queue.extend(adjacent_nodes)

    return visited

# Example usage
traversal_result = BFTraversal(graph, 'A')  # Start from a node that exists in the graph
print(f'BFT Order: {traversal_result}')
```

```python

#* Depth First Traverse
def DFTraversal(graph, start):
    if start not in graph:
        return "Start node not in graph"

    visited = []
    stack = [start]  # Use a stack for DFS

    while stack:
        node = stack.pop()  # Pop the last element (LIFO)

        if node not in visited:
            visited.append(node)
            # Push adjacent nodes to the stack in reverse order
            # to maintain the correct traversal order
            adjacent_nodes = graph.get(node, [])
            stack.extend(reversed(adjacent_nodes))

    return visited

# Example usage
start = 'A'  # Start from a node that exists in the graph
traversal_result = DFTraversal(graph, start)
print(f'DFT Order: {traversal_result}')
```
