## Searching : Visiting a purpose of a GOAL

```python

#? Graph


graph = {
    'S': ['B', 'D', 'A'],
    'A': ['C'],
    'B': ['D'],
    'C': ['G', 'D'],
    'D': ['G']
}
```

```python

#* Breadth First Search
def BFS(graph, start, goal):
    visited = []
    queue = [[start]]

    while queue:
        path = queue.pop(0)
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            return path

        adjacent_nodes = graph.get(node, [])
        for node2 in adjacent_nodes:
            new_path = path.copy()
            new_path.append(node2)
            queue.append(new_path)

    return None


start = 'S'
goal = 'G'
solution = BFS(graph, start, goal)

display((f'Solution is: {solution}'))
```

```python

#* Depth First Search
def DFS(graph, start, goal):
    visited = []
    stack = [[start]]

    while stack:
        path = stack.pop()
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            return path

        adjacent_nodes = graph.get(node, [])
        for node2 in adjacent_nodes:
            new_path = path.copy()
            new_path.append(node2)
            stack.append(new_path)

    return None

start = 'S'
goal = 'G'
solution = DFS(graph, start, goal)

print('Solution is', solution)
```

```python
def UCS(graph_with_cost, start, goal):
    visited = []
    queue = [[(start, 0)]]

    while queue:
        queue.sort(key=lambda x: path_cost(x))  # sorting by cost
        path = queue.pop(0)  # choosing least cost
        node = path[-1][0]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            return path
        else:
            adjacent_nodes = graph_with_cost.get(node, [])
            for (node2, cost) in adjacent_nodes:
                new_path = path.copy()
                new_path.append((node2, cost))
                queue.append(new_path)

def path_cost(path):
    return sum(cost for (node, cost) in path), len(path)


solution = ucs(graph_with_cost, 'S', 'G')
print('Solution is', solution)
print('Cost of Solution is', path_cost(solution)[0])
```
